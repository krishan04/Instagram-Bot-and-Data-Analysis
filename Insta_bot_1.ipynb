Problem 2: Type for “food” in search

```python
def Search(string):
    User_Handle=[]
    wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div'))).click()
    search_input = wait.until(EC.presence_of_element_located((By.XPATH,'//input[@placeholder="Search"]')))
    #search_input.clear()
    search_input.send_keys(string)
    userId = wait.until(EC.presence_of_all_elements_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[2]/div/a')))

    for i in userId:
        html_content=i.get_attribute('outerHTML')
        User_Link = BeautifulSoup(html_content,'html.parser').a['href']
        Get_User_Handle = User_Link.split('/')[1]
        if Get_User_Handle!='explore':
            User_Handle.append(Get_User_Handle)
            
    return User_Handle
```

```python
#searching the given string and printing the shown user handle
user_handles = Search('food')
for i in user_handles:
    print(i)
```

Problem 3 : Searching and Opening a profile

```python
def Search_and_Open_Profile(string):
    try:
        wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div'))).click()
        search_input = wait.until(EC.presence_of_element_located((By.XPATH,'//input[@placeholder="Search"]')))
        search_input.send_keys(string)
        time.sleep(5)
        handles = driver.find_elements('tag name', "a")
        handles[11].click()
    except:
        print("Something went Wrong")
```

```python
#Searching and open the given user handle
Search_and_Open_Profile('So delhi')
```

Problem 4 : Follow/Unfollow given handle - 

```python
def Follow():
    try:
        button = driver.find_element('class name','_ap30')
        if button.text == 'Follow':
            button.click()
            print("Started following 'So Delhi'")
        else:
            print('Already following \'So Delhi\'')
    except:
        print("Something went Wromg")
```

```python
def UnFollow():
    try:
        button = driver.find_element('class name','_ap30')
        if button.text == 'Following':
            button.click()
            print('Unfollowed "So delhi"')
            time.sleep(1)
            driver.find_element('xpath','/html/body/div[6]/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div/div[8]/div[1]').click()
        else:
            print('Not following \'So Delhi\'')
    except:
        print("Something went Wromg")
```

```python
#search and follow the user handle the unfollow it
Search_and_Open_Profile("So Delhi")
time.sleep(2)
Follow()
time.sleep(4)
UnFollow()
```

5. Like/Unlike posts

```python
def Liking_the_post():
    try:
        #Opening the first post
        driver.find_element('class name','_aagu').click()
        for i in range(30):
            time.sleep(2)
            e = driver.find_element('class name','xp7jhwk')
            GetText = BeautifulSoup(e.get_attribute('innerHTML'),'html.parser').svg['aria-label']
            if GetText == 'Like':
                #linking the post
                e.click()
                time.sleep(1)
                e.click()
                time.sleep(1)
            else:
                print('Already Liked')
            #next post
            driver.find_element('class name','_aaqg').click()
        # driver.back()
        driver.find_element("class name",'x1eu8d0j ').click()
    except:
        print("Something Went Wrong!")
```

```python
def Unliking_the_post():
    try:
        #Opening the first post
        driver.find_element('class name','_aagu').click()
        for i in range(30):
            time.sleep(2)
            e = driver.find_element('class name','xp7jhwk')
            GetText = BeautifulSoup(e.get_attribute('innerHTML'),'html.parser').svg['aria-label']
            if GetText == 'Unlike':
                #Unlinking the post
                e.click()
                time.sleep(1)
            else:
                print('Already Unliked')
            #next post
            driver.find_element('class name','_aaqg').click()
        #driver.back()
        driver.find_element("class name",'x1eu8d0j ').click()
    except:
        print("Something Went Wrong!")
```

```python
def Extract_followers():
    try:
        f = driver.find_elements('tag name','li')
        f[1].click()
        time.sleep(3)
        d = driver.find_element('class name','_aano')
        for i in range(50):
            driver.execute_script('arguments[0].scrollTop=arguments[0].scrollHeight',d)
            time.sleep(2)
        print("Scroll Down Complete")
        names = driver.find_elements('class name','x1dm5mii')
        a=1
        followers = []
        for i in names[:500]:
            followers.append(i.text.split("\n")[0])
        close = driver.find_element('class name','_abl-')
        close.click()
        return followers
    except:
        print("Something went wrong")
```
